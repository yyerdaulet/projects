```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
#Link to data : "andonians/random-linear-regression"
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
X_train,y_train = train['x'].to_frame(),train['y'].to_frame()
X_test,y_test = test['x'].to_frame(),test['y'].to_frame()
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
```

```python
print(f"Coef : {model.coef_}")
print(f"intercept : {model.intercept_}")
print(f"MSE : {mse}")
print(f"R2 : {r2}")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X_test,y_test,color='blue',label='Ndata')
plt.plot(X_test,y_pred,color='red',linewidth=1,label='Predicted')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression")
plt.legend()
plt.grid(True)
plt.show()
```
